```python
from cpscheduler.environment import SchedulingCPEnv, PrecedenceConstraint, Makespan
from cpscheduler.environment.constraints import ResourceCapacityConstraint

from cpscheduler.environment.instances import read_rcpsp_instance

import pandas as pd
import numpy as np
```

```python
path = 'instances/rcpsp/1kAD_2001.rcp'
```

```python
read_rcpsp_instance(path)
```

```python
instance, metadata = read_rcpsp_instance(path)
```

```python
f"{60.0:g}"
```

```python
env = SchedulingCPEnv(instance, 'duration')
env.add_constraint(
    ResourceCapacityConstraint(
        env.tasks,
        ['resource_0', 'resource_1', 'resource_2'],
        metadata['resource_capacities']
    )
)

env.add_constraint(
    PrecedenceConstraint,
    None,
    metadata['precedence_tasks']
)

env.set_objective(Makespan)
```

```python
env.reset()
```

```python
_, order, _, _ = env.get_cp_solution(solver='cplex')
```

```python
env.get_cp_solution()
```

```python
env.reset()
```

```python
env.step(order)
```
